```python
import pandas as pd 
```

```python
df = pd.read_csv('taxi.csv')
df.info()
```

```python
df.head()
```

```python
# Добавляем для каждой записи в какой день недели она была создана 
df['pickup_day'] = pd.to_datetime(df['pickup_datetime'])
df['pickup_day'] = df['pickup_day'].dt.day_name()
# Фильтруем по выходным
df_weekend = df[(df['pickup_day'] == 'Saturday') | (df['pickup_day'] == 'Sunday')]

df_weekend.head()
```

```python
# Добавляем для каждой записи в какой месяц недели она была создана 
df['pickup_month'] = pd.to_datetime(df['pickup_datetime'])
df['pickup_month'] = df['pickup_month'].dt.month_name()
# Группируем по месяцам 
df.groupby('pickup_month').agg(
    total_orders = ('id', 'count')
).reset_index()
```
